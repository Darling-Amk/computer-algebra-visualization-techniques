```python



ABC = [[120,55,50],[120,85,45],[220,80,40],[300,50,55],[30,27,15],[50,30,15],[50,30,13],[30,20,10],    ]
fig, axs = plt.subplots(nrows= 8 , ncols= 1 )

fig.set_figheight(8*4)
fig.set_figwidth(4)
for i in range(8):
  A,B,C = ABC[i]
  t = np.linspace(0,2*np.pi*(B/np.gcd(A, B)))
  axs[i].axhline(y=0, color='r')
  axs[i].axvline(x=0, color='g')
  axs[i].title.set_text(f"A={A} B={B} C={C}")
  axs[i].plot(
      (A-B)*np.cos(t)+C*np.cos(A*t/B),
      (A-B)*np.sin(t)-C*np.sin(A*t/B),
      )
  
  
```

##2)

```python


fig = plt.figure()
d = 10
ax = plt.axes(xlim=(-3*d, 3*d), ylim=(-3*d, 3*d))
line, = ax.plot([], [], lw=3)

 
#  .
def init():
    #   .
    line.set_data([], [])
    return line,

#  
def animate(C):
    T = np.linspace(0,2*np.pi*(B/np.gcd(A, B)))
    # x, y   
    x = [(A-B) * cos(t) + C * cos(A*t/B) for t in T]
    y = [(A-B) * sin(t) - C * sin(A*t/B) for t in T]
    #        x, y
    line.set_data(x, y)
    return line,

 
anim = FuncAnimation(fig, 
                     animate,
                     init_func=init,
                     frames=100,
                     interval=20,
                     blit=True,
                     )
```
