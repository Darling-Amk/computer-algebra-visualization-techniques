```python
n = 100
b = .3
fig, axs = plt.subplots(nrows= 6 , ncols= 1 )
fig.set_figheight(6*4)
fig.set_figwidth(4)
def Xn(n,A):
  ans = [b]
  for i in range(n):
    ans.append(4*A*ans[-1]*(1-ans[-1]))
  return ans

for j,A in enumerate((.2,.3,.5,.7,.8,.9)):
  X = [i for i in range(n+1)]
  axs[j].plot(X,Xn(n,A))

```

```python
r = 2.5
R = 4
N = 8
w = 5
```

```python
ax = plt.gca()
c = plt.Circle((0,0),radius=R/2,color="blue",linewidth=2, fill=False)
plt.gca().add_artist(c)
c = plt.Circle((0,0),radius=r/2,color="yellow",linewidth=2, fill=False)
plt.gca ().add_artist(c)

plt.figure(figsize=(9,9))  

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

fi = pi/N

X = []
Y = []
a = 0
for i in range(1,2*N+3):
  if i%2==1:
    x = r/2*cos(a*pi/180)
    y = r/2*sin(a*pi/180)
    X.append(x*cos(fi)-y*sin(fi))
    Y.append(x*sin(fi)+y*cos(fi))
  else:
    x = R/2*cos(a*pi/180)
    y = R/2*sin(a*pi/180)
    X.append(x*cos(fi)-y*sin(fi))
    Y.append(x*sin(fi)+y*cos(fi))
  a+=180/N


ax.plot(X,Y,"r")
```

```python
fig = plt.figure(figsize=(9, 9))

c = plt.Circle((0,0),radius=R/2,color="blue",linewidth=2, fill=False)
plt.gca().add_artist(c)
c = plt.Circle((0,0),radius=r/2,color="yellow",linewidth=2, fill=False)
plt.gca ().add_artist(c)


ax = plt.axes(xlim=(-R/2, R/2), ylim=(-R/2, R/2))
line, = ax.plot([], [], lw=3)

# функция анимации
def animate(t):
    fi=-w*pi/N+ t*pi/180
    X = []
    Y = []
    a = 0
    for i in range(1,2*N+3):
      if i%2==1:
        x = r/2*cos(a*pi/180)
        y = r/2*sin(a*pi/180)
        X.append(x*cos(fi)-y*sin(fi))
        Y.append(x*sin(fi)+y*cos(fi))
      else:
        x = R/2*cos(a*pi/180)
        y = R/2*sin(a*pi/180)
        X.append(x*cos(fi)-y*sin(fi))
        Y.append(x*sin(fi)+y*cos(fi))
      a+=180/N

    line.set_data(X, Y)
    return line,
# Функция инициализации.
def init():
    # создение пустого графа.
    line.set_data([], [])
    return line,


 
anim = FuncAnimation(fig, 
                     animate,
                     init_func=init,
                     frames=360,
                     interval=abs(w),
                     blit=True,
                     )
```
