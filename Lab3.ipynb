```python
# X
def f(t):
    return (2*t*(np.cos(t)**2)-0.1 )

fig = plt.figure()
ax = fig.add_subplot( projection='3d')

X = np.arange(-2, 1, 0.25)
Y = np.arange(0, 1, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X)

for fi in np.arange(0,2*np.pi,np.pi/720):
    X_=X
    Y_=Y*np.cos(fi)+Z*np.sin(fi)
    Z_ = -Y*np.sin(fi)+Z*np.cos(fi)
    X = X_
    Y= Y_
    Z = Z_
    ax.plot_surface(X, Y, Z,color='r',antialiased=True)


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
# Z
def f(t):
    return (2*t*(np.cos(t)**2)-0.1 )

fig = plt.figure()
ax = fig.add_subplot( projection='3d')

X = np.arange(-2, 1, 0.25)
Y = np.arange(0, 1, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X)

for fi in np.arange(0,2*np.pi,np.pi/720):
    X_=X*np.cos(fi)-Y*np.sin(fi)
    Y_=X*np.sin(fi)+Y*np.cos(fi)
    Z_ = Z

    X = X_
    Y= Y_
    Z = Z_
    ax.plot_surface(X, Y, Z,color='r',antialiased=True)


ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```
